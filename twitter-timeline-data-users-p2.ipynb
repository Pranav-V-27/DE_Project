```python
import pandas as pd
import tweepy
import csv
from datetime import datetime, timedelta
from tqdm.auto import tqdm

# Twitter API credentials

consumer_key = 'XXXXX'
consumer_secret = 'XXXXX'
access_token = 'XXXXX'
access_token_secret = 'XXXXX'




# consumer_key = 'XXXXX'
# consumer_secret = XXXXX'
# access_token = 'XXXXX'
# access_token_secret = 'XXXXX'


# Authenticate with Twitter API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
def get_timeline_sn(user):
    try:
        timeline = []
        for tweet in tweepy.Cursor(api.user_timeline, screen_name=user, count = 200).items(3000):
            timeline.append(tweet._json)
    except:
        timeline = []
    return timeline

def get_timeline_id(user):
    try:
        timeline = []
        for tweet in tweepy.Cursor(api.user_timeline, user_id=user, count = 200).items(3000):
            timeline.append(tweet._json)
    except:
        timeline = []
    return timeline
```

```python
def columns_by_part(part):
    cols = main_users.columns.tolist()
    temp = cols[(part - 1)*10:(part*10)]
    return temp
def concat_ids(cols):
    p = main_users[cols]
    ids = p.values.reshape(-1)
    return list(ids)
```

```python
PART = 2
print(columns_by_part(PART))
id_list = concat_ids(columns_by_part(PART))
print(len(id_list))
```

```python
timelines = []
```

```python
for i in tqdm(id_list):
    timelines.append(get_timeline_id(i))
```

```python
import json
json.dump(timelines, open(f'user_timelines_{PART}.json', 'w'), indent = 1)
```
