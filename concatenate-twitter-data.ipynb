```python
main_dfs = []
for i in file_paths_with_sizes:
    print(i)
    df = pd.read_csv(i[0])
    data = []
    for k,x in tqdm(df.iterrows(), total = len(df)):
        try:
            data.append((x['created_at'], eval(x['user'])['name'],eval(x['user'])['description'], x['text']))
        except:
            pass
    main_dfs.append(pd.DataFrame(data))
    del df
    del data
    gc.collect()
main_df = pd.concat(main_dfs).reset_index(drop = True)
main_df.columns = ['timestamp','user','bio','tweettext']
main_df.to_csv('main_df_users.csv', index = False)
```
