```python
def get_activity_timeline_sn(user):
    try:
        timeline = []
        act = api.get_favorites(screen_name = user, count = 200)
        for tweet in act:
            timeline.append(tweet._json)
    except:
        timeline = []
    return timeline

def get_activity_timeline_id(user):
    try:
        timeline = []
        act = api.get_favorites(user_id = user, count = 200)
        for tweet in act:
            timeline.append(tweet._json)
    except:
        timeline = []
    return timeline
```

```python
import random
def columns_by_part(part):
    cols = main_users.columns.tolist()
    temp = cols[(part - 1)*10:(part*10)]
    return temp
def concat_ids(cols):
    p = main_users[cols]
    ids = p.values.reshape(-1)
    return list(ids)
def select_random(ids):
    random.seed(100)
    return random.sample(ids, 3000)
```

```python
PART = 6
print(columns_by_part(PART))
id_list = concat_ids(columns_by_part(PART))
rid_list = select_random(id_list)
print(len(rid_list))
```

```python
timelines = []
```

```python
for i in tqdm(rid_list):
    timelines.append(get_activity_timeline_id(i))
```

```python
import json
json.dump(timelines, open(f'activity_timelines_{PART}.json', 'w'), indent = 1)
```
